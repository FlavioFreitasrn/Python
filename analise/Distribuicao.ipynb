```python
#Criando um variavel com 1000 elements para simular um distribuição normal
dados = stats.norm.rvs(size=1000)
```

```python
plt.hist(dados, bins = 20)
```

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```
